Графік розподілу зарплати


```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Зарплата на місяць'].dropna(), bins=20, kde=True, color='skyblue')
plt.title('Розподіл зарплати на місяць')
plt.xlabel('Зарплата на місяць')
plt.ylabel('Частота')
plt.show()
```

Графік взаємозв'язку між досвідом та зарплатою


```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Досвід роботи на поточному місці', y='Зарплата на місяць', data=df, hue='Мова програмування', palette='viridis', s=100)
plt.title('Взаємозв\'язок між досвідом роботи та зарплатою')
plt.xlabel('Досвід роботи (в роках)')
plt.ylabel('Зарплата на місяць')
plt.show()
```

Кругова діаграма розподілу мов програмування
